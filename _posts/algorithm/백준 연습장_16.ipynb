```
# 피보나치 수열...
# 다양한 해법 그 중 simple recursion(재귀)
# https://devdoggo.netlify.com/post/alg_ds/challenges/fibonacci_solutions/

n = int(input())

def fibonacci(n):
  if n < 2:
    return n

  return fibonacci(n - 1) + fibonacci(n - 2)

fibonacci(n)
```

## 그러나 이러한 방법은 저장을 하지 못하므로 여러 번 반복해야 한다는 맹점이 존재....

```
# 피보나치
# cache를 만들어 recursion을 저장하자(Memoization)

memo = {1:1, 2:1}
def fibonacci(n):
  if n == 0:
    return 0 
  if n not in memo:
    memo[n] = fibonacci(n - 1) + fibonacci(n - 2) # 재귀는 맞음. 메모값이 생성될때까지 저장하며 반복
  return memo[n]

```

```
# 피보나치 faast multiplication
# memoization은 하향식 저장인 반면 해당 코드는 상향식 저장
n = int(input())

def fibonacci(n):
  a, b = 1, 0
  for i in range(n):
    a, b = b, a + b # 동시에 처리해야하기 때문에 같은 문장에 써야함

  return b

print(fibonacci(n))
```
