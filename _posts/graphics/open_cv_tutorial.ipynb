### Open CV 사용법
cv.imread(file_nmae, flag): 이미지를 numpy객체로 만드는 함수
* file_name: 읽고자 하는 이미지 파일
* flag: 이미지를 읽는 방법 설정
 * IMREAD_COLOR:이미지를 COLOR로 읽고 투명한 부분은 무시
 * IMREAD_GRAYSCALE: 이미지를 Grayscale로 읽기
 * IMREAD_UNCHANGED: 이미지를 Color로 읽고, 투명한 부분도 읽기(alpha)
* return vlaue: numpy 객체(행, 열, 색상: 기본 BGR)**RGB가 아님에 주의**

cv2.imshow(title, image): 특정한 이미지를 화면에 출력함
* title: 윈도우 창의 제목
* image: 출력할 이미지 객체
> 파이참에서만 사용.... 코랩은 matplotlib사용

cv2.mwrite(file_name, image): 특정한 이미지를 파일로 저장하는 함수
* file_name: 저장할 이미지 이름
* image: 저장할 이미지 객체

cv2.waitKey(time): 키보드 입력을 처리하는 함수
* time: 입력대기 시간(무한: 0)
* return: 사용자가 입력한 Ascii Code

cv2.destroyALLWindows(): 화면닫기

```
import cv2
import matplotlib.pyplot as plt

img_basic = cv2.imread('asdf.jpg', cv2.IMREAD_COLOR) # 이미지를 컬러로 읽음
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_BGR2RGB)) # 4줄에서 읽어온 이미지를 RGB로 바꾸어서 plot해줌
# matplotlib을 쓰려면 어쩔수 없는 과정
plt.show() 


img_basic = cv2.cvtColor(img_basic, cv2.COLOR_BGR2GRAY) # 4줄에서 읽어온 이미지를 GRAY로바꾸어줌
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_GRAY2RGB)) # GRAY에서 GRB로 바꾸어 plot함
plt.show()
```

## 이미지 연산


```
import cv2
image = cv2.imread('asdf.jpg')

print('shape', image.shape) # 픽셀 수 확인
print('size', image.size) # 이미지 크기 확인

# 픽셀에 직접적으로 접근할 수 있음
px = image[100, 100]

# px에 저장되는 것은 BGR값임을 확인 -----> RGB가 아님에 주의!!!!!
print('pixel', px)

# px의 R값만 출력하기
print(px[2])
```
