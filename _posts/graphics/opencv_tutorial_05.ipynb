## filtering
이미지에 커널을 적용하여 이미지를 흐리게 처리할 수 있음 -> 노이즈 제거 가능

#### convolution 계산
window를 통해 convolution계산을 하여 필터링을 수행할 수 있음(가중치(커널의 각각에 대한 값) 적용하고 평균을 구하는 방식...)

#### kernel의 종류
* Basic Kernel: 모두 같은 값을 가짐
* Gaussian Kernel: 중간에 높은 값을 가지는 커널

```
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('asdf.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

size = 4 # 4x4
kernel = np.ones((size, size), np.float32) / (size ** 2) # 4x4 사이즈의 16분의 1kernel을 생성
print(kernel)

dst = cv2.filter2D(image, -1, kernel)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('asdf.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

dst = cv2.blur(image, (4, 4)) # 자동 커널 생성
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

```
import cv2

image = cv2.imread('asdf.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

# kernel_size: 홀수
dst = cv2.GaussianBlur(image, (5, 5), 0)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```
